**Data Analysis for Weather Monitoring**

```python
temp = np.random.randint(15,41,size=(5,7))
```

```python
temp
```

```python

for x in temp:
  sum = 0
  for y in x:
    sum+=y
  avg = sum/7
  print(avg)
```

```python
for x in temp:
  max = 0
  for y in x:
    if y>max:
      max=y
  print(max)
```

```python
temp.sort()
```

```python
temp
```

```python
temp.ravel()
```

**Ecommerce Sales Data**

```python
sales = np.random.randint(50,201,size=(3,7))
```

```python
sales
```

```python
for x in sales:
  sum = 0
  for y in x:
    sum+=y
  print(sum)
```

```python
for x in sales:
    total = 0
    for y in x:
        total += y
    print(total)
```

```python
new_sales = sales.copy()
new_sales[new_sales > 150] = 150
print(new_sales)
```
